Python画图库Seaborn中的kdeplot()函数可以绘制“核密度图”，用来估计数据的密度函数，从而展现数据样本本身的分布特征， kdeplot()函数语法为：kdeplot(data, shade=False, ax=None)，其中：

data -- 输入数据。

shade -- 如果为True，则在KDE曲线下方的区域中着色。

ax -- 如果提供，则在此ax上绘图。


Python画图库Seaborn中的regplot()函数可以绘制“回归关系图”，用来展现数据之间的回归关系， regplot()函数语法为：regplot(x, y, data=None, axis=None)，其中：

x, y, -- 输入变量。如果为字符串，则表示为列名，对应data中的某一列

data -- 输入数据

ax -- 如果提供画布ax，则在ax上绘图


下面我们利用kdeplot()函数绘制总消费金额的“核密度图”，展示其分布情况。 同时利用regplot()函数绘制总消费笔数和总消费金额两特征的回归关系图。

- 请使用regplot()函数在画布ax2中画出总消费笔数和总消费金额两特征的回归关系图，且设置x轴为总消费笔数，y轴为总消费金额。
- 图形的其他参数保持默认值。


[](http://)由图得到此数据集中，总消费金额为长尾分布，同时消费笔数与金额呈正相关关系。 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
data = pd.read_table('dataset.txt', encoding='gbk', sep=' ')

# 建立画布ax1和ax2,及设置图像大小，设置subplots()函数中参数为(1,2)表示两画图呈一行两列
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(16, 5))

# 在画布ax1中画出总消费金额的核密度图
sns.kdeplot(data['cashTotalAmt'], ax=ax1)
ax1.set_title('总消费金额的核密度图')

# 在画布ax2中画出总消费笔数和总消费金额的回归关系图
sns.regplot(x='transTotalCnt', y='transTotalAmt', data=data, ax=ax2)

ax2.set_title('总消费笔数和总消费金额的回归关系图')

# 显示图像
plt.show()


```

## 4.15 查看存在缺失值的特征

sort_values()函数可以将Series变量中的元素进行排序，函数语法为：sort_values(axis=0, ascending=True)，其中：
- axis：选择按行排序或者按列排序。0表示按行排序，1表示按列排序。
- ascending： 选择升降序，True表示按升序排列，False则为降序。

count()是pandas工具包中的自带函数，用于计算数据集中各个特征的有效值数量（各个特征中不为0的行数）。

## 实训任务

- 请用isnull()和sum()函数计算特征缺失值个数。
- 请用sort_values()函数将na_counts大于0的部分进行降序排序，将排序结果赋予变量missing_value。
 

通过查看存在缺失值的特征，初步可知数据集中有6个字段存在缺失值，分别是：婚姻状况maritalStatus、教育程度education、身份验证idVerify、性别sex、三要素验证threeVerify和是否违约Default。 


```python
import pandas as pd
data = pd.read_table('dataset.txt',encoding='gbk',sep=' ')
# 计算特征缺失值个数
na_counts = data.isnull().sum()

# 将na_counts取大于0的部分进行降序排序
missing_value = na_counts[na_counts>0].sort_values(ascending=False)

# 查看存在缺失值的特征
print(missing_value)
```

打印所有离散型特征的value_counts()的结果，根据每列的实际场景判断是否有异常值。

无犯罪记录isCrime字段的正常取值只为0、1，分别代表无/有犯罪记录，但存在6行取值为2，应进行相应的处理，对于离散型变量较常用的方法是采取众数填补。


## 实训任务

- 请将isCrime列中值为2的数用众数进行替换。

```python
import pandas as pd
data = pd.read_table('dataset1.txt', sep=' ')

# 计算isCrime列的众数
mode_isCrime = data['isCrime'].mode()[0]

# 将isCrime列中值为2的数用众数进行替换
data['isCrime'] = data['isCrime'].replace(2, mode_isCrime)

# 查看处理后的数据情况
print(data['isCrime'].value_counts())
```
